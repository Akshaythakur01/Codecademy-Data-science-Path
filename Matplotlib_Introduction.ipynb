```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
import numpy as np
```

```
x=np.arange(0,10)
y=np.arange(11,21)
```

```
a=np.arange(40,50)
b=np.arange(50,60)
```

```
##Scatter plot

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```
y=x*x
##plt plot
plt.plot(x,y,'r*--',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```
#Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

```
x=np.arange(1,11)
y=3*x+5
plt.title('Matplotlib')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
```

```
##Plotting Sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')
##
plt.plot(x,y)
```

```
## subplots of sine wave form
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

## Set up a subplot grid h=2,w1
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sine')


plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')
```

```
##Bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,align='center')
plt.title('Bar Graph')
plt.xlabel('x axis ')
plt.ylabel('y axis ')
```

```
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
```

```
##Box plot using Matplotlib
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

```
##Pie chart
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```
